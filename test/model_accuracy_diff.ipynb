```python
import pandas as pd

# 두 CSV 파일을 각각 읽어옵니다
file1 = "./result/kmmlu_category_metrics_기본.csv"
file2 = "./result/kmmlu_category_metrics_건축학.csv"

data1 = pd.read_csv(file1)
data2 = pd.read_csv(file2)

# 'accuracy'와 'category' 열이 각각의 CSV 파일에 포함되어 있다고 가정합니다
accuracy1 = round(data1['accuracy'], 3)
accuracy2 = round(data2['accuracy'], 3)
category = data1['category']  # 두 파일의 category는 동일하다고 가정

# 두 파일의 accuracy 값을 비교하는 데이터프레임 생성
comparison = pd.DataFrame({
    "Category": category,
    "Accuracy_기본_모델": accuracy1,
    "Accuracy_파인튜닝_모델": accuracy2,
    "Accuracy_차이": accuracy2 - accuracy1,
})

# Category를 인덱스로 설정
comparison.set_index('Category', inplace=True)
```

```python
comparison
```

```python
# 평균 정확도 차이, 정답 수 차이 출력
average_diff = comparison['Accuracy_차이'].mean()
print(f"\n평균 정확도 차이: {average_diff:.3f}")
print(f"정답 수 차이: {data2['correct_answers'].sum()-data1['correct_answers'].sum()}개")
print(f"기본 전체 평균: {(data1['correct_answers'].sum()/data1['total_questions'].sum())*100:.2f}%")
print(f"전체 평균: {(data2['correct_answers'].sum()/data2['total_questions'].sum())*100:.2f}%")
```
